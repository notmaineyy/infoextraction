## Dependency Parser

```python
import spacy

# load nlp pipeline
nlp = spacy.load('en_core_web_trf')

# process document: incl. tokenization, pos-tagging, dependency parsing, ...
doc = nlp('Elizabeth died on Sept 8 at her Scottish summer home, Balmoral Castle.')

# access dependency relations
for tok in doc:
    print(tok.head, '--->', tok, tok.dep_, sep="\t")
```

```python
import spacy
from spacy import displacy

options = {"compact": True}
displacy.render(doc, style='dep',jupyter=True,options=options)
```
